```python
#to calculate the values at the most recent date 
def most_recent_total(data):
    most_recent_date = data['date'].max()
    most_recent_tdata = data[data['date'] == most_recent_date]
    
    return most_recent_tdata
```

```python
most_recent_date = cont_total_death_df ['date'].max()
# most_recent_date
most_recent_td = cont_total_death_df[cont_total_death_df ['date'] == most_recent_date]
most_recent_td
```

```python
print(most_recent_total(cont_total_death_df).reset_index())
```

```python
melted_df = most_recent_td.melt(var_name='continent',value_name='total_no_deaths')
melted_df
```

```python
melted_df_cont = melted_df.iloc[1:, :].reset_index().drop("index",axis=1)
```

```python
melted_df_cont
```

```python
sum_world_td = melted_df_cont["total_no_deaths"].sum()
```

```python
sum_world_td
```

```python
melted_df_cont['percentage'] = melted_df_cont['total_no_deaths'].apply(lambda x: x / sum_world_td * 100)
```
